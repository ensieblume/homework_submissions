```python
# Dependencies and Setup
import pandas as pd
import numpy as np
np.random.seed(0)

# Raw data file|
file_to_load = "purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
total_players_df = purchase_data["SN"].value_counts()

#create the dataframe
frame_df = pd.DataFrame({"Total Players": [total_players_df.count()]})
frame_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data["Item ID"].value_counts()
unique_items.count()
average_price = purchase_data["Price"].mean()
purchases_num = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()
summary_purchase_total_df = pd.DataFrame({"Number of Unique Items": [unique_items.count()],
                         "Average Price": ['${:,.2f}'.format(average_price)],
                         "Number of Purchases": [purchases_num],
                         "Total Revenue": ['${:,.2f}'.format(total_revenue)]})
summary_purchase_total_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Create a new dataframe with one record per unique user, columns are Gender and sum of purchases
gender_demographics_df = purchase_data.groupby(["SN", "Gender"], as_index=False).agg({'Item ID': 'count', 'Price': 'sum','Age': 'mean'})
gender_demographics_df = gender_demographics_df.rename(columns={'Item ID':'NumberOfPurchases', 'Price':'TotalSpent'})
gender_demographics_df.head(10)
```

```python
#result1_df["Price"] = result1_df["Price"].map('${:,.2f}'.format)
gender_share = (gender_demographics_df["Gender"].value_counts(normalize=True)).map('{:,.2%}'.format)
gender_share
gender_count = gender_demographics_df["Gender"].value_counts()
gender_count
```

```python
gender_summary = pd.DataFrame({
    "Percentage of Players": gender_share,
    "Total Count": gender_count
})

gender_summary
```

```python
gender_total_count = purchase_data.groupby("Gender").count()["SN"]
numberOfPlayers = total_players_df.count()

gender_percentage = gender_total_count / numberOfPlayers
gender_percentage
```

```python
numberOfPlayers = total_players_df.count()
gender_summary_df = gender_demographics_df.groupby(["Gender"]).agg({'NumberOfPurchases': lambda x: 100*np.count_nonzero(x)/numberOfPlayers,
                                                                    'SN': 'count'})
gender_summary_df = gender_summary_df.rename(columns={'SN':'Total Count', 'NumberOfPurchases':'Percentage of Players'})
gender_summary_df.round(2)
```


## Purchasing Analysis (Gender)

```python
# tried two different ways to calculate purchase analysis, and preferred no lambda solution, but I am still keeping them for futre reference 
```

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
result1_df = purchase_data.groupby(["Gender"]).agg({'Purchase ID': 'count', 
                                                    'Price':['mean','sum'],
                                                   })                                                    
                                                             
result2_df = gender_demographics_df.groupby(['Gender']).agg({'TotalSpent': 'mean',
                                                            }) 
                                                            
result_summary_df = pd.concat([result1_df, result2_df], axis=1, sort=False)
result_summary_df.round(2)

result_summary_df.columns
```

```python
result_summary_df.rename(columns = {('Purchase ID', 'count'):'Purchase Count',
                                                       ("Price", "mean"):'Average Purchase Price',
                                                       ('Price', 'sum'): 'Total Purchase Value',
                                                        'TotalSpent': 'Avg Purchase Total per Person'
                                                      }, inplace=True)
# result_summary_df.round(2)
result_summary_df['Avg Purchase Total per Person'] = result_summary_df['Avg Purchase Total per Person'].map('${:.2f}'.format)
result_summary_df['Average Purchase Price'] = result_summary_df['Average Purchase Price'].map('${:.2f}'.format)
result_summary_df['Total Purchase Value'] = result_summary_df['Total Purchase Value'].map('${:.2f}'.format)
result_summary_df
```

```python
# Place the data series into a new column inside of the DataFrame
gender_demographics_df["Age Group"] = pd.cut(gender_demographics_df["Age"], age_bins, labels=group_names)
gender_demographics_df.head()
```

```python
age_share = gender_demographics_df["Age Group"].value_counts(normalize=True)
print(age_share)
age_count = gender_demographics_df["Age Group"].value_counts()
print(age_count)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a new dataframe with one record per unique Item ID and Item Name, columns are Gender and sum of purchases
most_popular_items_df = purchase_data.groupby(["Item ID", "Item Name"], as_index=False).agg({'SN': 'count', 'Price': 'sum'})
most_popular_items_df = most_popular_items_df.rename(columns={'SN':'NumberOfPurchases', 'Price':'TotalSpent'})
most_popular_items_df.head(10)
```

```python
most_popular_df = purchase_data.groupby(["Item ID", "Item Name"]).agg({'Purchase ID': 'count', 
                                                    'Price':'mean',
                                                   })         
                                                  
                                                             
most_popular2_df = most_popular_items_df.groupby(["Item ID", "Item Name"]).agg({'TotalSpent': 'sum',
                                                            }) 
                                                            
most_popular_summary_df = pd.concat([most_popular_df, most_popular2_df], axis=1, sort=False)
descending_most_popular_summary_df = most_popular_summary_df.sort_values(by='Purchase ID', ascending=False)
descending_most_popular_summary_df.head()
```
